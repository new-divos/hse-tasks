# Построение таблицы истинности

```python
import numpy as np
import pandas as pd
```

Определение класса бита для упрощения вычислений:

```python
class Bit:
    
    def __init__(self, value: int):
        self.__value = 1 if value else 0
        
    def __int__(self):
        return self.__value
        
    def __str__(self):
        return str(self.__value)
    
    def __invert__(self):
        return Bit(1 - self.__value)
    
    def __and__(self, other):
        if isinstance(other, Bit):
            return Bit(1 if self.__value and other.__value else 0)
        else:
            return Bit(0)
        
    def __or__(self, other):
        if isinstance(other, Bit):
            return Bit(1 if self.__value or other.__value else 0)
        else:
            return Bit(self.__value)
        
        
    def __xor__(self, other):
        if isinstance(other, Bit):
            return Bit(1 if self.__value != other.__value else 0)
        else:
            return Bit(self.__value)
```

Создать датафрейм для хранения результатов вычисления таблицы истинности

```python
data = np.array(
    [
        [0] * 8 + [1] * 8,
        ([0] * 4 + [1] * 4) * 2,
        [0, 0, 1, 1] * 4,
        [0, 1] * 8
    ],
    dtype=np.int
)

tt = pd.DataFrame(
    data=data.T,
    columns=['x1', 'x2', 'x3', 'x4']
)
```

Определить функцию для вычисления значения зависимой переменной

$$
y = \overline{x_1 \overline{x_3 x_4}} + \overline{\overline{x_2}x_4(x_2 + \overline{x_3} + x_4)} + \overline{x_3}\,\overline{x_4} + (x_2 + x_1 \overline{x_2 x_3}\,\overline{x_4} + x_3)\overline{x_3}.
$$

```python
def f(x1: Bit, x2: Bit, x3: Bit, x4: Bit) -> Bit:
    return ~(x1 & ~(x3 & x4)) | ~(~x2 & x4 & (x2 | ~x3 | x4)) | \
        (~x3 & ~x4) | (x2 | (x1 & ~(x2 & x3) & ~x4) | x3) & ~x3
```

Определить значение зависимой переменной

```python
tt['y'] = np.vectorize(
    lambda x1, x2, x3, x4: int(f(Bit(x1), Bit(x2), Bit(x3), Bit(x4)))
)(tt.x1, tt.x2, tt.x3, tt.x4)
```
